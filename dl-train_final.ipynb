```python
class validation_first_dataread():
    def __init__(self, dataset_path, table, transform = None):
        self.table = table.to_numpy()
        self.path = dataset_path
        self.transform = transform
            
    def __getitem__(self, index):
        img = Image.open(self.path + self.table[index][0])
        img = img.convert('RGB')
        if self.transform is not None:
            img = self.transform(img)
        else:
            img = ToTensor()(img)
            
        pseudo_label = self.table[index, 1]
        if (pseudo_label == 3):
            pseudo_label = 2
        if (pseudo_label == 4):
            pseudo_label = 3
        if (pseudo_label == 5):
            pseudo_label = 0
        label = torch.from_numpy(np.array([[pseudo_label]]))
        
        return img, label.squeeze()
            
    def __len__(self):
        #print(self.table)
        return(self.table.shape[0])
```

```python
from sklearn.metrics import confusion_matrix
def comp_accuracy(train_table, test_table, model, model_gm, model_sb):
    model.eval()
    model_gm.eval()
    model_sb.eval()
    y_true = []
    y_pred = []
    for index, row in test_table.iterrows():
        #print(row['Category'])
        y_true.append(row['Category'])
        if row['Category'] == 3 or row['Category'] == 2:
            img = Image.open('/kaggle/input/intro-dl/data/train/' + row['Id'])
            img = img.convert('RGB')
            img = transform(img)
            img = img.unsqueeze(0)
            
            output = model_gm(img.to(device))
            _, predicted = torch.max(output, 1)
            local_pr = predicted[0].item() + 2
            #print(local_pr)
            y_pred.append(local_pr)
        else:
            if row['Category'] == 0 or row['Category'] == 5:
                img = Image.open('/kaggle/input/intro-dl/data/train/' + row['Id'])
                img = img.convert('RGB')
                img = transform(img)
                img = img.unsqueeze(0)
            
                output = model_sb(img.to(device))
                _, predicted = torch.max(output, 1)
                local_pr = predicted[0].item()
                if (local_pr == 1):
                    local_pr += 4
                #print(local_pr)
                y_pred.append(local_pr)
            else:
                img = Image.open('/kaggle/input/intro-dl/data/train/' + row['Id'])
                img = img.convert('RGB')
                img = transform(img)
                img = img.unsqueeze(0)
            
                output = model(img.to(device))
                _, predicted = torch.max(output, 1)
                local_pr = predicted[0].item()
                if (local_pr == 3):
                    local_pr += 1
                y_pred.append(local_pr)
    
    y_pred = np.array(y_pred)
    y_true = np.array(y_true)
    y_res = y_pred - y_true
    
    T = [y_res == 0]
    print(T)
    N = len(test_table)
    corr = np.array([y_pred == y_true]).sum()
    print("acc: {:.4f}".format(corr/N))
    print(confusion_matrix(y_true, y_pred))
```
